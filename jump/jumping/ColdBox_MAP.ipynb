```python
%%cpp -d
double fillPoly(TFitResultPtr r, double WP)
{

double x[1] = { WP };
double err[1];  // error on the function at point x0

r->GetConfidenceIntervals(1, 1, 1, x, err, 0.683, false);
cout << " function value at " << x[0] << " => " << " +/- " << err[0] << endl;
return err[0];    
}
```
